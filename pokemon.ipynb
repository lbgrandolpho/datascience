```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pokedex.csv')
```

```python
pokemontype = []
pokemontype_number = []

for index, row in df.iterrows():
    if row['species'] in pokemontype:
        pokemontype_number[pokemontype.index(row['species'])] += 1
    else:
        pokemontype.append(row['species'])
        pokemontype_number.append(1)
```

```python
pokemontype1 = np.array(pokemontype)
pokemontype_number1 = np.array(pokemontype_number)
```

```python
plt.figure(figsize=[25,10])
plt.bar(pokemontype1, pokemontype_number1, color='red')
plt.xticks(fontsize=5, rotation=45)
plt.show()
```
